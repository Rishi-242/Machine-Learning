```python
#26. Subset rows where an external array, foo, is True
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45]
})

# External Boolean array (must be same length as df)
foo = [True, False, True, False, True]

# Subset the rows where foo is True
subset = df[foo]

print(subset)
```

```python
#27. Subset rows where an external array, bar, is positive

import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [80, 65, 90, 70, 85]
})

# External array (e.g., a NumPy array or list of same length)
bar = np.array([1, -1, 3, 0, 2])  # positive: index 0, 2, 4

# Subset rows where bar > 0
subset = df[bar > 0]

print(subset)
```

```python
#28. Subset rows where foo is TRUE or bar is negative
import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [80, 65, 90, 70, 85]
})

# External arrays (must be same length as df)
foo = np.array([False, True, False, True, False])
bar = np.array([1, -1, 3, 0, -2])

# Subset rows where foo is True OR bar is negative
subset = df[(foo) | (bar < 0)]

print(subset)
```

```python
#29. Subset the rows where foo is not TRUE and bar is not negative
import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [80, 65, 90, 70, 85]
})

# External arrays
foo = np.array([False, True, False, True, False])
bar = np.array([1, -1, 3, 0, -2])

# Subset where foo is NOT True AND bar is NOT negative
subset = df[(~foo) & (bar >= 0)]

print(subset)

```

```python
#31. Subset by columns TransactionID and TransactionDate
df = pd.DataFrame({
    'TransactionID': [101, 102, 103],
    'TransactionDate': ['2025-04-20', '2025-04-21', '2025-04-22'],
    'Amount': [200, 300, 400]
})
subset = df[['TransactionID', 'TransactionDate']]
print(subset)
```

```python
#32. Subset by columns TransactionID and TransactionDate with logical operator
df = pd.DataFrame({
    'TransactionID': [101, 102, 103, 104],
    'TransactionDate': ['2025-04-20', '2025-04-21', '2025-04-22', '2025-04-23'],
    'Amount': [200, 300, 400, 500]
})
condition = (df['TransactionID'] > 102) & (df['TransactionDate'] > '2025-04-21')
subset = df.loc[condition, ['TransactionID', 'TransactionDate']]
print(subset)
```

```python
#33. Subset columns by a variable list of column names
df = pd.DataFrame({
    'TransactionID': [101, 102, 103],
    'TransactionDate': ['2025-04-20', '2025-04-21', '2025-04-22'],
    'Amount': [200, 300, 400],
    'CustomerID': [1, 2, 3]
})
columns_to_subset = ['TransactionID', 'Amount']
subset = df[columns_to_subset]
print(subset)
```

```python
#34. Subset columns excluding a variable list of column names

df = pd.DataFrame({
    'TransactionID': [101, 102, 103],
    'TransactionDate': ['2025-04-20', '2025-04-21', '2025-04-22'],
    'Amount': [200, 300, 400],
    'CustomerID': [1, 2, 3]
})
columns_to_exclude = ['Amount', 'CustomerID']
subset = df.drop(columns=columns_to_exclude)
print(subset)
```
