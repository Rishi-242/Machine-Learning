```python
#13. Call the above function to find min max of horsepower, length, fuel efficiency
df = pd.read_csv("car_sales.csv")
# Function to find min and max values of any column
def find_min_max(df,'Price in thousands'):
min_value = df['Price in thousands'].min()
max_value = df['Price in thousands'].max()
# Find the rows with the min and max values
min_row = df[df['Price in thousands'] == min_value]
max_row = df[df['Price in thousands'] == max_value]
return min_value, max_value, min_row, max_row
# Now call the function to find min and max for 'Horsepower'
hp_min, hp_max, hp_min_row, hp_max_row = find_min_max(df,"Horsepower")
# Print results
print("\nHorsepower - Min:", hp_min, "\n", hp_min_row)
print("\nHorsepower - Max:", hp_max, "\n", hp_max_row)
```

```python
#14. Plot histogram of continuous numerical variable : price,sales,hoursepower,fuelefficiency.
import matplotlib.pyplot as plt

# Assuming you're using pandas DataFrame
import pandas as pd

# Example DataFrame
# Replace this with your actual DataFrame
df = pd.DataFrame({
    'Price of thousands': [20, 25, 30, 15, 22, 28],
    'sales': [200, 220, 250, 180, 210, 240],
    'hoursepower': [150, 160, 170, 140, 155, 165],
    'fuelefficiency': [25, 27, 23, 30, 28, 26]
})

# List of continuous variables to plot
columns = ['Price of thousands', 'sales', 'hoursepower', 'fuelefficiency']

# Plotting histograms
plt.figure(figsize=(12, 8))
for i, col in enumerate(columns):
    plt.subplot(2, 2, i + 1)
    plt.hist(df[col], bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#15. Probability density distribution of continuous numerical variable-length
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create dummy length data
df = pd.DataFrame({
    'length': np.random.normal(loc=150, scale=10, size=100)  # 100 values around 150
})

# Plot KDE
sns.kdeplot(df['length'], shade=True, color='purple')
plt.title('Probability Density Distribution of Length')
plt.xlabel('Length')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
